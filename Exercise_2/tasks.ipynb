```python
import pandas as pd
# 1)Load the Air Travel Dataset using pd.read_csv()
url="https://people.sc.fsu.edu/~jburkardt/data/csv/airtravel.csv"
df= pd.read_csv(url)
```

```python
# 2) Display the shape and basic statistics
print(df.shape)
print(df.describe())
```

```python
# print top 10 rows
print(df.head(10))
```

```python
# 3) Plot the air travel in July across different years
import matplotlib.pyplot as plt

plt.plot(df.columns[1:], df.iloc[6, 1:], marker='o')
plt.title('Air Travel in July Across Different Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid()
plt.show()
```
