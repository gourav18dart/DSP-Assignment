```python
import pandas as pd
# 1) read the feedback.csv into a dataframe and display first 15 rows
df = pd.read_csv('feedback.csv')
print(df.head(15))
```

```python
# 2) display the column names, number of rows and any missing values
df.info()
```

```python
# 3) find how many total responses were collected
total_responses = len(df)
print("Total Responses Collected:", total_responses)
```

```python
# 4) save only the feedback with a rating of 4 or above into a new csv file
# creating a new column with name rating using mapping
rating_mapping = {
    "Excellent": 5,
    "Very Good": 4,
    "Good": 3,
    "Average": 2,
    "Poor": 1
}
df['Rating'] = df['How would you rate the overall course?'].map(rating_mapping)
filtered_df = df[df['Rating'] >= 4]
filtered_df.to_csv('filtered_feedback.csv', index=False)
```
